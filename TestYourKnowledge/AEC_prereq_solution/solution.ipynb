```python
import numpy as np
import control as ctl
import matplotlib.pyplot as plt
# %matplotlib widget
```

## Lead RC Circuit

```python
R1 = 10e3
R2 = 100e3
C1 = .160e-9

alpha = R1/(R1+R2)
omega_p = 1/(alpha*R2*C1)
omega_z = 1/(R2*C1)
```

```python
print(omega_z/6.28, omega_p/6.28)
```

```python
G = alpha * ctl.tf([1/omega_z, 1], [1/omega_p, 1])
```

```python
print(G)
```

```python
mag, phase, omega = ctl.bode(G, Hz=True)
plt.tight_layout()
```

```python
Timpulse, yimpulse_out = ctl.impulse_response(G)
plt.plot(Timpulse, yimpulse_out)
plt.grid(True)
```

```python
omega_p*(1-omega_p/omega_z)
```

```python
Tstep, ystep_out = ctl.step_response(G)
plt.plot(Tstep, ystep_out)
plt.grid(True)
```
