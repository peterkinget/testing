```python
import numpy as np
import control as ctl
import matplotlib.pyplot as plt
```

```python
R1 = 10e3
R2 = 100e3
C1 = 1.6e-9

alpha = R1/(R1+R2)
omega_p = 1/(alpha*R2*C1)
omega_z = 1/(R2*C1)
```

```python
G = alpha * ctl.tf([1/omega_z, 1], [1/omega_p, 1])
```

```python
print(G)
```

```python
mag, phase, omega = ctl.bode(G)
plt.tight_layout()
```
